```python
# Lagrangepunkt L2

mu_earth = 3.986e14
mu_sun = 1.327e20
omega = 1.991e-7
radius_orbit_earth = 1.496e11

r = np.linspace(radius_orbit_earth, 2e11, 1000)

def acceleration(r):
    a_earth = -mu_earth / pow(r - radius_orbit_earth, 2)
    a_sun = -mu_sun / pow(r, 2)
    a_centrifugal = pow(omega, 2) * radius_orbit_earth
    return a_earth + a_sun + a_centrifugal


fig, ax = plt.subplots(figsize=(7.5,5))
ax.set_xlabel(r'Abstand zur Sonne r in Mio km')
ax.set_ylabel(r'Acceleration(r)')
ax.grid()
ax.grid(which='minor', color = '#999999', alpha = 0.2, linestyle = '-')
ax.minorticks_on()

ax.plot(r / 1000000000, acceleration(r), color = 'xkcd:red pink', alpha = 0.5, linewidth = 1, label = r'a')

ax.legend()
ax.legend(loc="upper center")
plt.show()
```

```python
# f(x) = x  tanh(x/t)

def f(x, t):
    return x - np.tanh(x/t)

x = np.linspace(-1, 5, 1000)

fig, ax = plt.subplots(figsize=(7.5,5))
ax.set_xlabel(r'x')
ax.set_ylabel(r'f(x)')
ax.grid()
ax.grid(which='minor', color = '#999999', alpha = 0.2, linestyle = '-')
ax.minorticks_on()

ax.plot(x, f(x, 0.5), color = 'xkcd:red pink', alpha = 0.5, linewidth = 1, label = r'f(x) fr t = 0.5')

ax.legend()
ax.legend(loc="upper center")
plt.show()
```
